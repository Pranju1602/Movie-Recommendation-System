```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    words = text.split()
    words = [word for word in words if word not in stop_words]
    words = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(words)
    
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(
    n_neighbors=10,        # 1 movie khud hogi + 5 similar
    metric='cosine',
    algorithm='brute'
)

knn.fit(tfidf_matrix)
```

```python
import random
import numpy as np

def evaluate_model(knn, tfidf_matrix, df, top_k=5, sample_size=500):

    random.seed(42)      # 👈 Fix randomness
    np.random.seed(42)
    
    sample_indices = random.sample(range(len(df)), sample_size)
    
    scores = []
    
    for movie_index in sample_indices:
        
        input_genres = set(df.iloc[movie_index]['genres'].split())
        
        distances, indices = knn.kneighbors(
            tfidf_matrix[movie_index],
            n_neighbors=top_k + 1
        )
        
        match_count = 0
        total = 0
        
        for idx in indices[0]:
            if idx != movie_index:
                
                recommended_genres = set(df.iloc[idx]['genres'].split())
                
                if len(input_genres & recommended_genres) > 0:
                    match_count += 1
                
                total += 1
        
        if total > 0:
            scores.append(match_count / total)
    
    average_accuracy = np.mean(scores)
    
    return average_accuracy
```

```python
accuracy = evaluate_model(knn, tfidf_matrix, df, top_k=5, sample_size=500)

print("Model Accuracy:", round(accuracy * 100, 2), "%")
```

```python
import random
import numpy as np

def evaluate_model_tags(knn, tfidf_matrix, df, top_k=5, sample_size=100, threshold=0.3):
    """
    Evaluate recommender system using Jaccard similarity on tags.
    
    Parameters:
    - knn : trained KNN model
    - tfidf_matrix : TF-IDF feature matrix
    - df : dataframe containing movies with 'tags' column
    - top_k : number of recommendations
    - sample_size : number of movies to test
    - threshold : similarity threshold to count as correct
    
    Returns:
    - average accuracy score
    """
    
    # Fix randomness for reproducibility
    random.seed(42)
    np.random.seed(42)

    # Select fixed random movies
    sample_indices = random.sample(range(len(df)), sample_size)

    scores = []

    for movie_index in sample_indices:
        
        input_tags = set(str(df.iloc[movie_index]['tags']).split())

        distances, indices = knn.kneighbors(
            tfidf_matrix[movie_index],
            n_neighbors=top_k + 1
        )

        match_count = 0
        total = 0

        for idx in indices[0]:
            if idx != movie_index:
                
                recommended_tags = set(str(df.iloc[idx]['tags']).split())

                # Avoid division by zero
                union = input_tags | recommended_tags
                if len(union) == 0:
                    continue

                intersection = input_tags & recommended_tags
                similarity = len(intersection) / len(union)

                if similarity >= threshold:
                    match_count += 1

                total += 1

        if total > 0:
            scores.append(match_count / total)

    return np.mean(scores)
```

```python
accuracy = evaluate_model_tags(knn, tfidf_matrix, df, top_k=5, sample_size=500)

print("Model Accuracy:", round(accuracy * 100, 2), "%")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
def recommend(title,n=10):
    if title not in indices:
        return ['Movie not found']

    idx = indices[title]
    sim_score = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx = sim_score.argsort()[::-1][1:n+1]
    return df['title'].iloc[similar_idx]
    
```

```python
recommend('Avatar')
```
